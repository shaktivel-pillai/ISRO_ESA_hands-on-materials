```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import aspiics_colormaps

filename = "./tutorial_data/aspiics_bt_l3_1320404D0002_20250812T181424.fits"

with fits.open(filename) as hdul:
    data = hdul[0].data

data = np.array(data, dtype=float)

plt.figure(figsize=(6, 6))
plt.imshow(np.sqrt(data), origin="lower", cmap="bt", vmin=np.sqrt(1e-11), vmax=np.sqrt(1e-6))
plt.colorbar(label="sqrt(MSB)")
plt.title(filename)
plt.tight_layout()
plt.show()
```
