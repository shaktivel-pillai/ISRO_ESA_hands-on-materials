## **STEPS L2 DATA**

The **dataproducts** from STEPS are:

1. **Differential direction flux of ions for Parker spiral (PS) detector**,  
   e.g., `AL1_AST91_L2_PS_20240926_V01.cdf`

2. **Differential direction flux of ions for Earth pointing (EP) detector**,  
   e.g., `AL1_AST91_L2_EP_20240926_V01.cdf`

3. **Differential direction flux of ions for North pointing (NP) detector**,  
   e.g., `AL1_AST91_L2_NP_20240926_V01.cdf`

4. **Differential direction flux of ions for intermediate (IM) detector**,  
   e.g., `AL1_AST91_L2_IM_20240926_V01.cdf`

## **Merge the Data of All Directions Files and Read Required Variables**

```python
#%%################# Merging data of all direction of STEPS/AL1
import os
import numpy as np
from spacepy import pycdf  # For reading CDF files

# Define variable names for each direction
variable_names = {
    'PS': {
        'epoch': 'Epoch',
        'flux': 'PS_Outer_allspecies',
        'energy': 'Ebin_outer_allspecies'
    },
    'EP': {
        'epoch': 'Epoch',
        'flux': 'EP_Outer_allspecies',
        'energy': 'Ebin_outer_allspecies'
    },
    'NP': {
        'epoch': 'Epoch',
        'flux': 'NP_allspecies',
        'energy': 'Ebin_allspecies'
    },
    'IM': {
        'epoch': 'Epoch',
        'flux': 'IM_allspecies',
        'energy': 'Ebin_allspecies'
    }
}

def read_cdf_data(cdf_file, variables):
    """Read the specified variables from a CDF file."""
    with pycdf.CDF(cdf_file) as cdf_data:
        epoch_times = cdf_data[variables['epoch']][...]
        flux = cdf_data[variables['flux']][...]
        energy_bins = cdf_data[variables['energy']][...]
    return epoch_times, flux, energy_bins

def read_position_data(cdf_file):
    """Read pos_GSE from a single CDF file."""
    with pycdf.CDF(cdf_file) as cdf_data:
        pos_steps = cdf_data['pos_GSE'][...]
    return pos_steps

def process_direction_data(directory, direction):
    """Process all files for a given direction (e.g., 'PS') in the directory."""
    variables = variable_names[direction]
    all_epoch_times = []
    all_flux = []
    energy_bins = None

    for filename in sorted(os.listdir(directory)):
        if direction in filename and filename.endswith(".cdf"):
            file_path = os.path.join(directory, filename)
            epoch, flux, ebin = read_cdf_data(file_path, variables)

            all_epoch_times.append(epoch)
            all_flux.append(flux)
            if energy_bins is None:
                energy_bins = ebin

    # Merge data from multiple files
    all_epoch_times = np.concatenate(all_epoch_times)
    all_flux = np.concatenate(all_flux, axis=0)

    return all_epoch_times, all_flux, energy_bins

# Example usage
data_dir = r'E:\PhD\AL1SC Workshop\Data'  # Replace with your actual folder path

all_epochs = []
all_positions = []

# Loop over files and concatenate epochs and positions
for filename in sorted(os.listdir(data_dir)):
    if filename.endswith('.cdf'):
        file_path = os.path.join(data_dir, filename)
        with pycdf.CDF(file_path) as cdf:
            epochs = cdf['Epoch'][...]       # timestamps array for this file
            pos = cdf['pos_GSE'][...]        # position array for this file
            all_epochs.append(epochs)
            all_positions.append(pos)

# Concatenate along first axis to get full arrays aligned by index
all_epochs = np.concatenate(all_epochs)
all_positions = np.concatenate(all_positions)

# # Save them aligned for later usage
# np.save('epoch_steps.npy', all_epochs)
# np.save('pos_steps.npy', all_positions)


all_data = {}
for direction in ['PS', 'EP', 'NP', 'IM']:
    print(f"Processing direction: {direction}")
    epoch_times, flux, energy_bins = process_direction_data(data_dir, direction)
    all_data[direction] = {
        'epoch_times': epoch_times,
        'flux': flux,
        'energy_bins': energy_bins
    }

# Access merged data
epoch_times_psout = all_data['PS']['epoch_times']
flux_psout = all_data['PS']['flux']
energy_bins_psout = all_data['PS']['energy_bins']

epoch_times_epout = all_data['EP']['epoch_times']
flux_epout = all_data['EP']['flux']
energy_bins_epout = all_data['EP']['energy_bins']

epoch_times_im = all_data['IM']['epoch_times']
flux_im = all_data['IM']['flux']
energy_bins_im = all_data['IM']['energy_bins']

epoch_times_np = all_data['NP']['epoch_times']
flux_np = all_data['NP']['flux']
energy_bins_np = all_data['NP']['energy_bins']

# ----------------- Read and merge pos_GSE --------------------
# Collect pos_GSE from all CDF files
pos_steps_list = []

for filename in sorted(os.listdir(data_dir)):
    if filename.endswith(".cdf"):
        file_path = os.path.join(data_dir, filename)
        pos_data = read_position_data(file_path)
        pos_steps_list.append(pos_data)

# Concatenate all positions
pos_steps = np.concatenate(pos_steps_list, axis=0)

# # Save for future use
# np.save('pos_steps.npy', pos_steps)

# Print shape to confirm
print("Shape of pos_GSE (merged):", pos_steps.shape)

```

## **Plotting the Flux Time Series of STEPS-ASPEX L2 Data**

```python
#%% ########################## Plotting for Level 2 data of STEPS/AL1

import matplotlib.pyplot as plt
import pandas as pd  # Ensure pandas is imported

plt.figure(figsize=(12, 8))

# Define start and end date limits for the x-axis
start_date = '2024-09-01 00:00:00'
end_date = '2024-10-03 00:00:00'

# Convert start and end dates to datetime objects
start_datetime = pd.to_datetime(start_date)
end_datetime = pd.to_datetime(end_date)

# Loop through each energy bin to plot the corresponding flux
num_energy_bins = energy_bins_im.shape[0]  # Get the number of energy bins
selected_bins = 6  # Number of bins to plot

for i in range(selected_bins):
    # Define the energy range for the current bin
    lower_bound = energy_bins_im[i, 0]
    upper_bound = energy_bins_im[i, 1]
    energy_range = f"{lower_bound:.2f} - {upper_bound:.2f} MeV"
    
    # Plot flux for the current energy bin against time
    plt.plot(epoch_times_im, flux_im[:, i], '.', ms=3, label=energy_range)

plt.yscale('log')  # Set y-axis to log scale
plt.xlabel('Time (UTC)')
plt.ylabel('Ion flux \n (cm$\mathbf{^{2}}$-sec-sr-MeV)$\mathbf{^{-1}}$')
plt.xlim(start_datetime, end_datetime)
plt.title('Ion Flux Over Time for IM Direction')
plt.grid(True)

# Formatting the x-axis to show time correctly
plt.gcf().autofmt_xdate()  # Rotate date labels for better readability

# Add a legend to the plot
plt.legend(title='Energy Ranges')

# Show the plot
plt.tight_layout()
plt.show()
```

## Plotting of Flux Time Series for PS, EP, NP and IM Directions of STEPS L2 Data

```python
#%% ########## Multipanel Plot of L2 data of STEPS/AL1

import matplotlib.pyplot as plt
import pandas as pd  # Ensure pandas is imported

# Define a dictionary to access the data for each direction
directions_data = {
    'PS': {'epoch_times': epoch_times_psout, 'flux': flux_psout, 'energy_bins': energy_bins_psout},
    'EP': {'epoch_times': epoch_times_epout, 'flux': flux_epout, 'energy_bins': energy_bins_epout},
    'NP': {'epoch_times': epoch_times_np, 'flux': flux_np, 'energy_bins': energy_bins_np},
    'IM': {'epoch_times': epoch_times_im, 'flux': flux_im, 'energy_bins': energy_bins_im},
}

# Create a figure and 4 subplots (4 rows, 1 column)
fig, axs = plt.subplots(4, 1, figsize=(12, 9.5), sharex=True)

# Define start and end date limits for the x-axis 
start_date = '2024-09-01 00:00:00'
end_date = '2024-10-03 00:00:00'

# Convert start and end dates to datetime objects
start_datetime = pd.to_datetime(start_date)
end_datetime = pd.to_datetime(end_date)

# Loop over the directions and create a subplot for each
for idx, (direction, data) in enumerate(directions_data.items()):
    epoch_times = data['epoch_times']
    flux = data['flux']
    energy_bins = data['energy_bins']
    num_energy_bins = energy_bins.shape[0]  # Get the number of energy bins
    selected_energy_bins = 6

    for i in range(selected_energy_bins):
        # Define the energy range for the current bin
        lower_bound = energy_bins[i, 0]
        upper_bound = energy_bins[i, 1]
        energy_range = f"{lower_bound:.2f} - {upper_bound:.2f} MeV"
        
        # Plot flux for the current energy bin against time
        axs[idx].plot(epoch_times, flux[:, i], '.', ms=2, label=energy_range)
    
    axs[idx].set_yscale('log')  # Set y-axis to log scale
    axs[idx].set_ylabel('Ion flux \n (cm$\mathbf{^{2}}$-sec-sr-MeV)$\mathbf{^{-1}}$')
    axs[idx].set_title(f'Ion Flux Over Time for {direction} Direction')
    axs[idx].grid(True)
    axs[idx].legend(title='Energy Ranges', fontsize=8)

    # Set x-axis limits to the specified date range
    axs[idx].set_xlim(start_datetime, end_datetime)
    
    # Format x-axis
    ax = plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d\n%H:%M'))
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.grid(True)
    
    # Place the legend outside the plot area
    axs[idx].legend(title='Energy Ranges', fontsize=8, bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Place the legend outside the plot area with larger marker size and font
    axs[idx].legend(title='Energy Ranges', fontsize=9, markerscale=5, bbox_to_anchor=(1.05, 1), loc='upper left')



# Set x-axis label for the last subplot
axs[-1].set_xlabel('Time (UTC)')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()

```
