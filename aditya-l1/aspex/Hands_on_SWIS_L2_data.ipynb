## **SWIS L2 DATA**

There are **3 data products** from SWIS:

1. **Differential energy flux for THA-1**,  
   e.g., `AL1_ASW91_L2_TH1_20241231_UNP_9999_999999_V02.cdf`

2. **Differential energy flux for THA-2**,  
   e.g., `AL1_ASW91_L2_TH2_20241231_UNP_9999_999999_V02.cdf`

3. **Bulk parameters computed from the ion distribution function**,  
   e.g., `AL1_ASW91_L2_BLK_20241231_UNP_9999_999999_V02.cdf`

## **Now we will look at how to access and read ASPEX_SWIS CDF files, and perform data analysis**

## **Read In the Differential Energy Flux THA-1**

```python
# import necessary python libaries 
from spacepy import pycdf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the variables from energy flux THA-1
cdf_1=pycdf.CDF(r'D:/aditya_l1_workshop_data/AL1_ASW91_L2_TH1_20250403_UNP_9999_999999_V02.cdf')
```

```python
# global attributes for cdf 
print(cdf_1.attrs)
```

```python
# Variables in the cdf file
print(cdf_1)
```

## **Extracting variables for THA-1**

```python
#extracting variables from the cdf file
#Access data for a specific variable
time= cdf_1['epoch_for_cdf_mod'][...]                                # Read all entries of the Epoch variable
energy = cdf_1['energy_center_mod'][...]                             # Read all entries of the Energy variable
flux_main = cdf_1['integrated_flux_mod'][...]                        # Read all entries of the integrated flux variable
flux_s9 = cdf_1['integrated_flux_s9_mod'][...]                       # Read all entries of the integrated flux for sector 9 variable
flux_s10 = cdf_1['integrated_flux_s10_mod'][...]                     # Read all entries of the integrated flux for sector 10 variable
flux_s11 = cdf_1['integrated_flux_s11_mod'][...]                     # Read all entries of the integrated flux for sector 11 variable
x = cdf_1['spacecraft_xpos'][...]                                    # Read all entries of the spacecraft_postion x (GSE) variable
y = cdf_1['spacecraft_ypos'][...]                                    # Read all entries of the spacecraft_postion y (GSE) variable
z = cdf_1['spacecraft_zpos'][...]                                    # Read all entries of the spacecraft_postion z (GSE) variable
sun_angles = cdf_1['sun_angle_tha1'][...]                            # Read all entries of the sun_angles_tha1(GSE) variable
```

## **Plotting Flux Time Series at a specific energy**

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# Choose an energy index (e.g. center of the energy bins)
energy_index = 10  # change this to any valid index within energy.shape[1]

# Extract the time series of flux at that energy
flux_timeseries = flux_main[:, energy_index]
energy_value = energy[0, energy_index]  # assuming energy is constant over time

# Plotting
plt.figure(figsize=(12, 4))
plt.plot(time, flux_timeseries, label=f'Energy = {energy_value:.1f} eV')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.title(f'Flux Time Series at Energy = {energy_value:.1f} eV')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.xticks(rotation=30)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## **Read In the Differential Energy Flux THA-2**

```python
# extracting data for THA_2
cdf_2=pycdf.CDF(r'D:/aditya_l1_workshop_data/AL1_ASW91_L2_TH2_20250403_UNP_9999_999999_V02.cdf')
#list all the variables
print(cdf_2)
```

## **Extracting variables for THA-2**

```python
# Extract variables
time = cdf_2['epoch_for_cdf_mod'][...]                  # Shape: (17275,)
energy = cdf_2['energy_center_mod'][...]                # Shape: (17275, 50)
flux_main = cdf_2['integrated_flux_mod'][...]           # Shape: (17275, 50)
flux_s15 = cdf_2['integrated_flux_s15_mod'][...]        # Shape: (17275, 50)
flux_s16 = cdf_2['integrated_flux_s16_mod'][...]        # Shape: (17275, 50)
flux_s17 = cdf_2['integrated_flux_s17_mod'][...]        # Shape: (17275, 50)
flux_s18 = cdf_2['integrated_flux_s18_mod'][...]        # Shape: (17275, 50)
flux_s19 = cdf_2['integrated_flux_s19_mod'][...]        # Shape: (17275, 50)

xpos = cdf_2['spacecraft_xpos'][...]                    # Shape: (17275,)
ypos = cdf_2['spacecraft_ypos'][...]                    # Shape: (17275,)
zpos = cdf_2['spacecraft_zpos'][...]                    # Shape: (17275,)

sun_angle = cdf_2['sun_angle_tha2'][...]                # Shape: (17275, 32, 3)
```

## **2D Energy Spectrogram plotting for a sector**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.colors import LogNorm

# Ensure time is in matplotlib date format
time_num = mdates.date2num(time)  # time should be datetime-like
energy_vals = energy[0]           # assuming energy is constant over time
flux_vals = flux_main.T           # shape should be (energy, time)

# Plot using contourf
plt.figure(figsize=(12, 5))
levels = 50  # You can adjust this for resolution

contour = plt.contourf(time_num, energy_vals, flux_vals, levels=levels, cmap='jet', norm=LogNorm())

# Formatting
plt.yscale('log')  # Log scale on y-axis
plt.ylabel('Energy (eV)')
plt.xlabel('Time')
plt.title('Flux Time Series (Spectrogram)')
plt.colorbar(contour, label='Flux \n ev/sr-s-cm^2-ev')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

```

## **Read In Bulk Parameters**

```python
#%% Calling another CDF file for bulk plasma parameter extraction
# Load the CDF file for bulk parameters
cdf_bulk = pycdf.CDF(r'D:/aditya_l1_workshop_data/AL1_ASW91_L2_BLK_20250403_UNP_9999_999999_V02.cdf')
```

## **Extracting Bulk Parameters**

```python
# Extract relevant variables for plotting and analysis
time1 = cdf_bulk['epoch_for_cdf_mod'][...]       # Time variable
v_bulk = cdf_bulk['proton_bulk_speed'][...]      # Proton bulk speed (scalar magnitude)
vx = cdf_bulk['proton_xvelocity'][...]           # Proton velocity component in x-direction (GSE)
vy = cdf_bulk['proton_yvelocity'][...]           # Proton velocity component in y-direction
vz = cdf_bulk['proton_zvelocity'][...]           # Proton velocity component in z-direction
density = cdf_bulk['proton_density'][...]        # Proton number density
p_temp = cdf_bulk['proton_thermal'][...]         # Proton thermal temperature
```

## **Cleaning Bulk Parameters for any discrepency**

```python
#remove all the negative values for v_bulk , vx , v_bulk and density
v_bulk[v_bulk<0]=np.nan
vx=-1*vx
vx[(vx > 1000) | (vx < 0)] = np.nan
p_temp[p_temp<0]=np.nan
density[density<0]=np.nan
```

## **Time Series Plot for a given parameter**

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from spacepy import pycdf
from datetime import datetime

# Create the figure
plt.figure(figsize=(12, 4))

# Plot proton bulk speed
plt.plot(time1, v_bulk, color='tab:blue', linewidth=1.5, label='Proton Bulk Speed')

# Beautify the plot
plt.xlabel('Time', fontsize=12)
plt.ylabel('Bulk Speed (km/s)', fontsize=12)
plt.title('Proton Bulk Speed Over Time', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', fontsize=10)

# Format x-axis for better readability
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.xticks(rotation=30)
plt.tight_layout()

# Show the plot
plt.show()
```

## **Panel Plot of all the Bulk Parameters**



```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming time1 is datetime-like or numpy datetime64, if not convert accordingly

fig, axs = plt.subplots(6, 1, figsize=(12, 10), sharex=True)

# Plot each variable
axs[0].plot(time1, v_bulk, label='Bulk Speed')
axs[0].set_ylabel('Bulk Speed (km/s)')
axs[0].legend(loc='upper right')

axs[1].plot(time1, vx, label='Vx', color='C1')
axs[1].set_ylabel('Vx (km/s)')
axs[1].legend(loc='upper right')

axs[2].plot(time1, vy, label='Vy', color='C2')
axs[2].set_ylabel('Vy (km/s)')
axs[2].legend(loc='upper right')

axs[3].plot(time1, vz, label='Vz', color='C3')
axs[3].set_ylabel('Vz (km/s)')
axs[3].legend(loc='upper right')

axs[4].plot(time1, density, label='Density', color='C4')
axs[4].set_ylabel('Density (cm‚Åª¬≥)')
axs[4].legend(loc='upper right')

axs[5].plot(time1, p_temp, label='proton temperature')
axs[5].set_ylabel('Temperature(ev)')
axs[5].legend(loc='upper right')
# Format x-axis as time on bottom plot
axs[5].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d\n%H:%M'))
fig.autofmt_xdate()

# Remove vertical space between plots
plt.subplots_adjust(hspace=0)

# Set common x-label
axs[-1].set_xlim(time1[0], time1[-1])
axs[-1].set_xlabel('Time')

plt.show()
```
